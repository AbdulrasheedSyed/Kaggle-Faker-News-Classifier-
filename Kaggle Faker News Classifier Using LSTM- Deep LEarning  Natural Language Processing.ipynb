## Kaggle Faker News Classifier Using LSTM- Deep LEarning| Natural Language Processing

Dataset: http://www.kaggle.com/c/fake-news/data#

```python
import pandas as pd
```

```python
df= pd.read_csv('fake-news/train_fake-news.csv')
```

```python
df.head()
```

```python
## Drop nan values
df=df.dropna()
```

```python
# Get the Independent Features
X=df.drop('label', axis=1)
```

```python
# Get the dependent Features
y=df['label']
```

```python
X.shape
```

```python
y.shape
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
# Vocabulary size
voc_size=5000
```

# OneHot Reprasentation

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

## Data set Preprocessing

```python
from nltk.stem.porter import PorterStemmer # Stemming
```

```python
ps= PorterStemmer()
corpus=[]

for i in range(0, len(messages)):
    #print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words("english")]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Onehot Reprasentation¶

```python
onehot_repr=[one_hot(words, voc_size) for words in corpus]
onehot_repr
```

## Word Embedding Reprasentation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr, padding='pre',maxlen=sent_length)
print(embedded_docs)
```

## Creating Model

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size, embedding_vector_features,input_length=sent_length))
#model.add(Dropout(0.3))
model.add(LSTM(100))
#model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'] )
print(model.summary())
```

```python
import numpy as np
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X_final,y_final, test_size=0.33, random_state=42)
```

## Model Training

```python
## Finally Training
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=10, batch_size=64)
```

## Performance Matrix and Accuracy 

```python
y_pred=model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
